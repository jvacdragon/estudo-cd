```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt

base = pd.read_csv('./dados/trees.csv')
base.head()
```

```python
#histograma com kde (kernel density)
srn.histplot(base.Volume, color='blue', kde=True, stat='density', bins=10).set(title ='Volume')
```
