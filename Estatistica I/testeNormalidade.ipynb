```python
import scipy as sc
import matplotlib.pyplot as plt

#criando dados normalizados
dados = sc.stats.norm.rvs(size=10000)
#verificação por histPlot
plt.hist(dados, bins=30)
```

```python
#gerando um qqplot
fig, ax = plt.subplots()
sc.stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
#testede shapiro
#se p for maior que 0.05 então está em distribuição normal
sc.stats.shapiro(dados)
```
