```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn

base = pd.read_csv('./dados/trees.csv')
```

```python
#grafico de dispersao considerando volume e dispersao
plt.scatter(x=base.Girth,y=base.Volume,marker='.', facecolors='none', color='blue')
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#grafico de linha considerando o volume e a circunferencia 
plt.plot(base.Girth, base.Volume)
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#grafico de disperao com afastamento de dados
srn.regplot(x=base.Girth, y=base.Volume, data = base, fit_reg=False, x_jitter=0.3)
```

```python
plt.hist(base.Volume)
```

```python
#criando figura com os graficos
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(x=base.Girth, y=base.Volume, marker='.')
plt.subplot(2,2,2)
plt.scatter(x=base.Girth, y=base.Height, marker='.')
plt.subplot(2,2,3)
plt.scatter(x=base.Height, y=base.Volume, marker='.')
plt.subplot(2,2,4)
plt.hist(base.Volume)
```
