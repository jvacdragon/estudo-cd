```python
import pandas as pd 
import matplotlib.pyplot as plt

base = pd.read_csv('./dados/trees.csv')
base.head()
```

```python
#circunferencia e volume
plt.scatter(x=base.Girth, y=base.Volume, facecolors='none', color='blue', marker='.')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#dispersao circunferencia e altura
plt.scatter(x=base.Girth, y=base.Height, color='blue', facecolors='none', marker='.')
```
