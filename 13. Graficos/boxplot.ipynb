```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn

base = pd.read_csv('./dados/trees.csv')
base.head()
```

```python
plt.boxplot(base.Volume, vert=False, showfliers=False, notch=True, patch_artist=True)
plt.title('Arvores')
plt.xlabel('Volume')
```

```python
plt.boxplot(base, vert=False)
plt.title('Arvores')
plt.xlabel('Dados')
```

```python
plt.boxplot(base.Volume, vert=False, patch_artist=True)
plt.boxplot(base.Girth, vert=False, patch_artist=True)
plt.boxplot(base.Height, vert=False)
plt.title('Arvores')
plt.xlabel('dados')
```

```python
srn.boxplot(base, orient='h').set_title('Arvores')
```
