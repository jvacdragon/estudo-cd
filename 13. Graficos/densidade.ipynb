```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn

base = pd.read_csv('./dados/trees.csv')
```

```python
#histograma com 10 divisoes
plt.hist(base.iloc[:,1], bins=6)
```

```python
srn.histplot(base.iloc[:,1], kde=False, bins=6).set(title='Arvores')
```

```python
#kdeplot para densidade 
srn.kdeplot(base.iloc[:,1]).set(title='Arvores')
```

```python
srn.histplot(base.iloc[:,1], kde=True, bins=6).set(title='Arvores')
```
